## Learning Decision Tree Algorithm
Here is my project for learning the decision trees algorithm .its built on the iris dataset from sklearn library.

```python
from sklearn.datasets import load_iris
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,classification_report
from sklearn import tree
import graphviz
```

```python
iris=load_iris()
```

```python
iris.feature_names
```

```python
iris.target_names
```

```python
x=iris.data
y=iris.target
```

```python
df=pd.DataFrame(x,columns=iris.feature_names)
```

```python
df.head()
```

```python
df['species']=y
z=iris.target_names
z
```

```python
df.head()
```

```python
sp={
    0:'setosa',
    1:'versicolor',
    2:'virginica'
}
df['species']=(df['species']).map(sp)
```

```python
df.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
clf=DecisionTreeClassifier(max_depth=3,criterion='gini',random_state=42)
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
```

```python
print(f"the accuracy score is {accuracy_score(y_pred=y_pred,y_true=y_test)}")
```

```python
print(f"the classification report {classification_report(y_pred,y_test)}")
```

```python
tree.plot_tree(clf)
```
